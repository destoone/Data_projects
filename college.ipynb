# Visualizing earnings based on college majors

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
recent_grads= pd.read_csv("recent-grads.csv")
```

```python
recent_grads= recent_grads.dropna()
```

```python
cleaned_data_count= recent_grads.shape[0]
cleaned_data_count
```

```python
fig= plt.figure(figsize=(5,10))
col_x= ["Sample_size","Sample_size","Full_time","ShareWomen","Men","Women"]
col_y= ["Median","Unemployment_rate","Median","Unemployment_rate","Median","Median"]
for i in range(6):
    ax= recent_grads.plot(x= col_x[i],y= col_y[i], kind="scatter",)
    ax.set_title(col_x[i]+" "+"vs."+" "+col_y[i])
    ax.set_xlabel(col_x[i])
    ax.set_ylabel(col_y[i])
plt.show()
```

* More popular majors don't make more money because, sample size   column have weaks median salaries into the hight values.
* No the majore in subject that were majority female don't make     more money!
* No we haven't relation between full-time employees and   median salary!

```python
col_n= ["Sample_size","Employed","Full_time","Men","Women"]
fig= plt.figure(figsize=(5,15))
for i in range(5):
    ax= fig.add_subplot(5,1,i+1)
    ax.hist(recent_grads[col_n[i]],bins=25,range=(0,5000))
    ax.set_title(col_n[i])
plt.show()
```
